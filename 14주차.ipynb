```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
df = pd.read_csv('Most Popular Programming Languages.csv')

# 2. 'Month' 열 제거하고 나머지 열만 사용
language_data = df.drop(columns=['Month'])

# 3. 언어별 평균 인기도 계산
average_popularity = language_data.mean().sort_values(ascending=False)

# 4. 상위 10개 언어 추출
top10 = average_popularity.head(10)

# 5. 파이차트 시각화
plt.figure(figsize=(8, 8))
plt.pie(top10.values, labels=top10.index.str.replace(' Worldwide(%)', ''), autopct='%1.1f%%', startangle=140)
plt.title('Average Popularity of Top 10 Programming Languages')
plt.axis('equal')
plt.show()
```
