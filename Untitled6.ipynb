```python
import pandas as pd
import numpy as np

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
file = path+'vehicle_prod.csv'
df = pd.read_csv(file, index_col=0)

sums = df.select_dtypes(np.number).sum().rename('Total')
mean=df.select_dtypes(np.number).mean().rename('average')

df.loc['total']=sums
df.loc['average']=mean
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'
weather = pd.read_csv(weather_file, encoding='CP949')

# '일시'에서 월 추출
weather['month'] = pd.DatetimeIndex(weather['일시']).month
monthly_wind = weather.groupby('month')['평균풍속'].mean()

# 그래프 그리기
months = np.arange(1, 13)
plt.bar(months, monthly_wind, color='green')
plt.xlabel('month')
plt.ylabel('wind')
plt.xticks(months)
plt.title('monthly wind')
plt.show()
```
